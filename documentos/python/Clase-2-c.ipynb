### Grafico de firmas espectrales

Agregaremos en este caso dos comandos utiles para graficar. 

El primero `plt.figure(figsize(x,y))` nos deja cambiar el tamaño del grafico.

El segundo `plt.savefig("archivo.png")` nos deja guardar la figura a un archivo. 

Ingresamos luegos las firmas espectrales corregidas por *toa*, *cos*, *dos*, *6S* y la obtenida del USGS.

```python
# Valores de reflectancia
toa = [128,329,185,3568,1219,442] # TOA
cos = [329,185,3568,1219,442,128] # COS
dos = [185,3568,1219,442,128,329] # DOS 1%
sss = [3568,1219,442,128,329,185] # 6S
usg = [1219,442,128,329,185,3568] # USGS

# Valores de longitud de onda para landsat 8
x = [482,561,655,864,1608,2200]
```

```python
# Configuramos el tamaño del grafico
plt.figure(figsize=(8,6))
# Grafico
plt.plot(x,toa,linestyle = "-", marker='None', color="#e41a1c", label="TOA")
plt.plot(x,cos,linestyle = "-", marker='None', color="#377eb8", label="COS")
plt.plot(x,dos,linestyle = "-", marker='None', color="#4daf4a", label="DOS")
plt.plot(x,sss,linestyle = "-", marker='None', color="#984ea3", label="6S")
plt.plot(x,usg,linestyle = "-", marker='None', color="#ff7f00", label="USGS")


# Configuraciones de ploteo. 
# Cambiar bajo su propio riesgo
# Pongo el limite en x e y
plt.xlim(450,2300)
plt.ylim(0,4000)
# Incluyo la leyenda
plt.legend(numpoints=1)
# Pongo nombres a los ejes
plt.xlabel("Longitud de londa [nm]")
plt.ylabel("Relectancia [Arb.]")
# Pongo titulo
plt.title("Firmas espectrales")

# Guardo el grafico con la menor cantidad de borde
#plt.savefig("pepe.png", bbox_inches='tight')

# Muestro el grafico
plt.show()
```
